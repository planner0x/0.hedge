### Derivatives Volatility and Variance

##### Option Pricing and Hedging

* BSM in 1973 uncertainty of price of underlying can be modeled by geometric Brownian motion process and a stochastic differential equation
* $dS_t = \mu S_tdt + \sigma S_tdZ_t$
* The risk factor to the option is a stock index paying no dividends
* $S_t$: Index level at time t
* $\mu$: constant drift
* $\sigma$: instantaneous volatility
* $Z_t$: Standard Brownian motion

* In risk-neutral setting $\mu$ is replaced by constant risk-less short-rate r
* $dS_t = rS_tdt + \sigma S_tdZ_t$
* The index is assumed to be directly tradeable.
* There is an risk-less bond that is also assumed to be directly tradeable 
* $dB_t = rB_tdt$

* Price of vanilla European put option P with payoff $max[K - S_t,0]$ is determined by put-call parity:
    * $P_t = C_t - S_t + e^{-r(T-t)}K$
    
* You can replicate this option payoff with a portfolio of the underlying index and a risk-less bond
* The value of the option must equal the value of the payoff in order to avoid risk-less arbitrage
* You can also discount back the risk-neutral expectation of the option payoff using the risk-neutral short rate

* Dynamically trading the underlying index and a risk-less bond will allow us to delta hedge

* Assuming trading at each discrete time interval t, profit-Loss for a given time interval $t$: 
    * $PL_{\Delta t} \approx \frac {1}{2}\Gamma * \Delta S^2 + \Theta * \Delta t$
    * $\Gamma$: measures how option value changes with values in underlying index
    * $\Theta$: measures how option value changes with change in time $t$
* PL of a given delta hedge is the difference between theoretical (ex ante) priced variance and realized variance during a hedging period
* Variance futures and Variance Swaps (OTC) pay the holder the difference between a fixed variance strike and realized variance over period of time

##### Notions of Volatility and Variance

* Realized or Historical Volatility: Standard Deviations of log returns of financial time series
* Instantaneous Volatility: volatility factor of a diffusion process
* Implied Volatility: volatility that if you plug into BSM gives you the market observed price of option
 
* These volatility factors have squared counterparts called variance
* Annualized variance is the result of multiplying the variance by 252, the number of trading days in a year


##### Volatility and Variance as an asset class

* CBOE introduced the VIX in 1993
* Currently VIX represents a 30-day variance swap rate

* Trading in VIX futures began in 2004 
* The first options on the index were introduced in 2006

* In 2005 Eurex launched futures on VSTOXX volatility index, based on the EURO Stoxx 50
* In 2009, relaunched as Mini futures contract and attracted significant liquidity due to financial crisis
* After 2008, with volatility spiking, market participants became more sensitive to volatility risk 
* In 2010, options on VSTOXX index were created 
* In 2014, Eurex created variance futures contracts for EURO Stoxx 50

##### Vol of Vol Indexes

* VVIX and VVSTOXX
* VVSTOXX: 
    * based on VSTOXX realtime options prices, reflect the near to long-term expectation of volatility of volatility
    * Square root of implied variance across all options of a given time to expiration
* Hedge options written on the volatility index itself 
* Vol of vol is usually higher than volatility


##### Volatility and Variance Trading

##### Vol Trading

* Market expectation: VSTOXX represents 30-day implied vol average for out-of-the money options
    * It is market consensus of to be realized volatility over the next 30-days
* Non-tradeability: VSTOXX is not tradeable only derivatives can be traded
* mean-reverting in nature, no positive/negative drift
* negative correlation to its underlying index
* positive jumps
* VSTOXX is oftern higher than realized volatility

* Long VSTOXX future
    * used to hedge equity positions
    * also used to hedge a short realized vol strategy
* Short VSTOXX future
    * Can be used to speculate reversion to mean after volatility spike
    * Reduce vega exposure to long vega option portfolios
* Term Structure Arbitrage
    * If Contango: short front month and long the nearby future contract 
        * Possible if certain carries trades are available
* Relative value arbitrage
    * Trade against other variance and volatility instruments

* VSTOXX Specific
    * Since Options are European in nature, delta hedging is possible
    * Long OTM Calls: market crash insurance
    * Short ATM calls: If implied vol is high anf expecting mean reversion, you can write options and collect premium
    * Long ATM Straddle: When VVOl is high buy puy-call with same ATM strike and maturity yields profit when VSTOXX moves fast enought in one direction
    
* Generally, contango/backwardation trades, expectation vs realized vol arbitrage


##### Variance Trading

* Long Variance Future: Realized variance is greater than expected variance 
    * Hedge Equity positions or Short Vega options
* Short Variance Future: Realized Variance is lower than expected variance
    * Hedge a long Vega options
* Forward volatility/variance trading
    * Gain exposure to foward implied volatility, short the front month variance future and long the nearby month
