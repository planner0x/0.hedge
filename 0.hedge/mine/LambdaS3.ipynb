```python
import datetime
import boto3
today = datetime.date.today()

session = boto3.Session(profile_name='jgerard')
s3_session = session.client('s3')



bucket_name = 'vol-trading-option-chains'

# s3_bucket = s3_session.bucket(bucket_name)




def get_daily_options_chains(ticker):
    
    series = yf.Ticker(ticker)
    for i, expiration in enumerate(series.options):
        calls = series.option_chain(expiration).calls
        print('writing calls to local')
        calls.to_csv('/tmp/calls.csv')
        puts = series.option_chain(expiration).puts
        print('writing puts to local')
        puts.to_csv('/tmp/puts.csv')
        

        #####
        # Add objects to bucket
        print('uploading calls to bucket')
        s3_session.upload_file('/tmp/calls.csv', bucket_name, f'{ticker}/{today}/{i}/{expiration}/calls.csv')
        print('uploading puts to bucket')
        s3_session.upload_file('/tmp/puts.csv', bucket_name, f'{ticker}/{today}/{i}/{expiration}/puts.csv')
        
        
        
    
```
