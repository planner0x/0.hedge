#### Chapter 5: Complete Market Models

##### Black-Scholes-Merton
* BSM valuation of puts and calls
* Moneyness: the more an option moves ITM, the greater the value
* Options are worth less the more OTM
* Time to maturity: The more time to maturity the more the option is worth
* Short rate: Increase in short rate increases the value of the call option and decreases value of the put
* Volatility: Increase in volatility increases the value of the call option more than the put due to increase ITM outcomes

##### The Greeks in BSM
* How do each of these parameters affect the overall price of the option
* $\Delta$: First PDE wrt to the index value
* $\Gamma$: 2nd PDE wrt to index level
* $\Theta$: Negative first partial derivative wrt time
* $\rho$: first pde wrt the short rate r
* $V$: First PD wrt volatility

##### What can we learn about the Greeks
* Delta: The shorter the maturity, and the closer K is to the strike, the more delta is affected 
* Gamma: Short maturities, ATM affect gamma the most
* Theta: Short maturities, ATM affect theta negatively
* Rho: Longer maturities and moneyness shift the value of rho positively
* Vega: Increases with maturity and decreases in both directions ATM
